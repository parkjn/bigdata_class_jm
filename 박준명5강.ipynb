```python
import seaborn as sns
```

```python
iris=sns.load_dataset("iris")
```

```python
iris.head()
```

```python
import matplotlib.pyplot as plt
vals=iris['sepal_length'].values
sns.kdeplot(vals)
plt.title("kdeplot for sepal length")
plt.show()
```

```python
vals=iris['petal_length'].values
sns.kdeplot(vals)
plt.title("kdeplot for petal length")
plt.show()
```

```python
vals=iris['petal_length'].values
sns.distplot(vals,kde=True,rug=True)
plt.title("kdeplot for petal length")
plt.show()
```

```python
sns.displot(vals,kde=True,rug=True)
plt.title("displot for petal length")
plt.show()
```

```python
sns.histplot(vals,kde=True)
plt.title("hisplot for petal length")
plt.show()
```
